### `Problem 1`: Write a program that will give you in hand monthly salary after deduction on CTC - HRA(10%), DA(5%), PF(3%) and taxes deduction as below:

> Salary(Lakhs) : Tax(%)

*   Below 5 : 0%
*   5-10 : 10%
*   10-20 : 20%
*   aboove 20 : 30%

````python
# Write code here

ctc_input = float(input("Enter your CTC (in lakhs per annum): "))

# covert ctc from lakhs to rupees
ctc = ctc_input * 100000

# Calculate standard deduction
hra = ctc * 0.1
da = ctc * 0.05
pf = ctc * 0.03

# calculate total deduction
total_deduction = hra + da + pf

# calculate taxable salary after deducton
taxable_salary = ctc - total_deduction

# Applying taxes according to slab
if ctc_input < 5:
  tax_rate = 0
elif ctc_input <=10:
  tax_rate = 10
elif ctc_input <=20:
  tax_rate = 20
else:
  tax_rate = 30

# calculate taxes on salary
tax = (tax_rate/100) * taxable_salary

# Annual salary
in_hand_annual_salary = taxable_salary - tax

# Monthly salary
in_hand_monthly_salary = in_hand_annual_salary/12

print(f"\nðŸ‘‰ Your monthly in-hand salary is: â‚¹{in_hand_monthly_salary}")
````

### `Problem 6` - Find the factorial of a given number.

Write a program to use the loop to find the factorial of a given number.

The factorial (symbol: `!`) means to multiply all whole numbers from the chosen number down to 1.

For example: calculate the factorial of 5

```bash
5! = 5 Ã— 4 Ã— 3 Ã— 2 Ã— 1 = 120
```

Output:

```bash

```

````python
# Write code here

num = int(input("enter the number = "))

fact = 1
for i in range (num,0,-1):
  fact = fact * i

print(f"factorial of {num} is :",fact)
````

### `Problem 8`: Take a user input as integer N. Find out the sum from 1 to N. If any number if divisible by 5, then skip that number. And if the sum is greater than 300, don't need to calculate the sum further more. Print the final result. And don't use for loop to solve this problem.

**Example 1:**

`Input:`

```bash
30
```

`Output:`

```bash
276
```

````python
# Write code here

N = int(input("enter the integer = "))
sum = 0
i = 1

while i <= N +1:
  if i % 5 == 0:
    i += 1
    continue

  sum += i

  if sum > 300:
    sum -= i
    break

  print(i, end = " ")
  i+=1

print ("")
print(sum)
````

### `Problem 9`: Write a program that keeps on accepting a number from the user until the user enters Zero. Display the sum and average of all the numbers.

````python
# Write code here

num = int(input("enter number = "))
sum = 0
counter = 0
while True:

  if num == 0:
    break
    print("You enter the invalid value")

  print(num)
  sum += num
  num = int(input("enter number = "))
  counter += 1

print("Sum = ",sum)
print("counter = ",counter)
average = sum//counter
print("Average =",average)

````

###`Problem 9`: Write a program which will find all such numbers which are divisible by 7 but are not a multiple of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a comma-separated sequence on a single line.

````python
# Write code here

i = 2000
L = []
while i <= 3200+1:
  if i % 7 == 0 and i % 5 != 0:
    L.append(str(i))
  i += 1

print(",".join(L))
````

###`Problem 10`: Write a program, which will find all such numbers between 1000 and 3000 (both included) such that each digit of the number is an even number. The numbers obtained should be printed in a space-separated sequence on a single line.



````python
# Write code here

L = []

for i in range (1000,3001):
  flag = True
  curr = i

  while curr > 0:
    last = curr % 10
    if last % 2 != 0:
      flag = False
      break
    curr = curr//10
  if flag:
    L.append(str(i))
print(",".join(L))
````

###`Problem 11`: A robot moves in a plane starting from the original point (0,0). The robot can move toward UP, DOWN, LEFT and RIGHT with a given steps.
The trace of robot movement is shown as the following:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
> The numbers after the direction are steps.

> `!` means robot stop there.

**Please write a program to compute the distance from current position after a sequence of movement and original point.**

*If the distance is a float, then just print the nearest integer.*

Example:

`Input`:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
`Output`:
```
2
```

````python
# Write code here

pos = [0,0]
while True:
  s = input("enter the movement = ")
  if s == "!":
    break
  direction = s.split()[0]
  magnitude = int(s.split()[1])

  if direction == "UP":
    pos[1] = pos[1] + magnitude
  elif direction == "DOWN":
    pos[1] = pos[1] - magnitude
  elif direction == "LEFT":
    pos[0] = pos[0] - magnitude
  elif direction == "RIGHT":
    pos[0] = pos[0] + magnitude
  else:
    pass

print(round((pos[0]**2 + pos[1]**2)**0.5,1))
````
